```python
import numpy as np

# Tạo hai tensor 3D ngẫu nhiên
A = np.random.rand(3, 4, 5)
B = np.random.rand(5, 5, 7)

print(A , "\n")
print(B)
# Nhân tensor A và B trên trục thứ 2 của A và trục thứ 0 của B
C = np.tensordot(A, B, axes=([2], [1]))
print("\n ================================= \n")
print(C.shape)  # Kết quả sẽ là (3, 4, 6, 7)
print(C)
```
